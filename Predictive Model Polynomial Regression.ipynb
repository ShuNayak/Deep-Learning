Maybe the data does not have a linear relationship, maybe it is a curved line.
The more degrees of polynimial isn't always better, this may lead to overfitting, doesn't do a good job at predicting values.

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

np.random.seed(12)
age = np.random.normal(55,10,50)
wealth = np.random.normal(1000000,4000,50)/age

plt.scatter(age, wealth)
```

```python
#numpy has a ployfit function
x = np.array(age)
y = np.array(wealth)

polyplot = np.poly1d(np.polyfit(x,y,4))
xSet = np.linspace(10,80,50)
plt.scatter(age, wealth)
plt.plot(xSet, polyplot(xSet),c='g')
```

```python
from sklearn.metrics import r2_score
r2 = r2_score(y, polyplot(xSet))
print (r2)
```

Here the r2 value is negative hence it fits worse than a horizontal line.
