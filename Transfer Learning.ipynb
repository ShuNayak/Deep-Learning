The idea is to use pre trained models that are already out there for use. For example if you are trying to do image classification then there are pretrained models like ResNet, VGG that we can import and use. Use them as is or tune them for your application. This is called transfer learning, so basically we are transfering an exisiting model developed by someone to your problem.

```python
#let us use the ResNet50 that is used for image classification, trained on imagenet data set to quickly classify new images
#let us load the image of the space needle, I like it.
from IPython.display import Image
Image(filename='Space_Needle.jpg')
```

```python
from tensorflow.keras.applications.resnet50 import ResNet50
from tensorflow.keras.preprocessing import image
from tensorflow.keras.applications.resnet50 import preprocess_input, decode_predictions
import numpy as np
```

```python
imagePath = 'Space_Needle.jpg'
img = image.load_img(imagePath, target_size=(224,224))
x = image.img_to_array(img)
x = np.expand_dims(x, axis=0)
x = preprocess_input(x)
```

```python
model = ResNet50(weights='imagenet')
```

```python
preds = model.predict(x)
print('Predicted: ',decode_predictions(preds,top=3)[0])
```

```python
def classifyImage(im):
    display(Image(filename=im))
    img = image.load_img(im,target_size=(224,224))
    x = image.img_to_array(img)
    x = np.expand_dims(x,axis=0)
    x = preprocess_input(x)
    preds = model.predict(x)
    print('predicted: ',decode_predictions(preds,top=3)[0])
```

```python
classifyImage('Indian-Breakfast.jpg')
```

```python
classifyImage('hritik.jpg')
```

```python
classifyImage('tiger.jpg')
```

```python
classifyImage('tom.jpg')
```
