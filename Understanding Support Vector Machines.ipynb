```python
import numpy as np
import pandas as pd

#create a fake cluster for n people in k clusters, age/no of cars
def createFakeCluster(N, k):
    np.random.seed(24)
    pointsInCluster = float (N)/k
    x = []
    y = []
    for i in range(k):
        ageCentroid = np.random.uniform(20.0, 70.0)
        carCentroid = np.random.uniform(10.0,50.0)
        for j in range(int(pointsInCluster)):
            x.append([np.random.normal(ageCentroid, 2.0), np.random.normal(carCentroid, 5.0)])
            y.append(i)
    x = np.array(x)
    y = np.array(y)
    return x,y


            
```

```python
%matplotlib inline
import matplotlib.pylab as plt
from sklearn.preprocessing import MinMaxScaler

(X,y) = createFakeCluster(100,6)
plt.figure(figsize=(8,6))
plt.scatter(X[:,0], X[:,1], c=y.astype(np.float))
plt.show()

scaling = MinMaxScaler(feature_range=(-1,1)).fit(X)
X = scaling.transform(X)

plt.figure(figsize=(10,10))
plt.scatter(X[:,0],X[:,1],c=y.astype(np.float))
plt.show()




```

```python
from sklearn import svm, datasets
c = 1.0
svc = svm.SVC(kernel='linear', C=c).fit(X,y)
```

```python
print(svc.predict(scaling.transform([[20,56]])))
```
